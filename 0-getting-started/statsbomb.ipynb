```python
from mplsoccer import Sbopen

parser = Sbopen()

# competition data
df_competition = parser.competition()
df_competition.head()
```

```python
# match data
df_match = parser.match(competition_id=9, season_id=281)
df_match.info()
```

```python
# get match_id from first match
first_match_id = df_match.iloc[0]['match_id']

# get lineup data for first match
df_lineup = parser.lineup(first_match_id)
df_lineup

# get all Belgian players
df_belgian_players = df_lineup[df_lineup['country_name'] == 'Belgium']
df_belgian_players
```

```python
df_event, df_related, df_freeze, df_tactics = parser.event(first_match_id)
df_event.info()

# show events where "dribble_nutmeg" is not null
nutmeg_events = df_event[df_event['dribble_nutmeg'].notna()]

# make a list of all player names that nutmegged someone
nutmeg_players = nutmeg_events['player_name'].unique()
nutmeg_players
```

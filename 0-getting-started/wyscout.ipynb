```python
import pathlib
import os
import pandas as pd
import json
```

```python
# COMPETITIONS DATA
notebook_dir = pathlib.Path().absolute()
competitions_path = os.path.join(notebook_dir.parent, "wyscout-data", "competitions.json")

with open(competitions_path) as f:
    competitions = json.load(f)

df_competitions = pd.DataFrame(competitions)
df_competitions
```

```python
# MATCHES DATA
matches_path = os.path.join(notebook_dir.parent, "wyscout-data", "matches", "matches_England.json")

with open(matches_path) as f:
    matches = json.load(f)

df_matches = pd.DataFrame(matches)
df_matches.head()
```

```python
# PLAYERS DATA
players_path = os.path.join(notebook_dir.parent, "wyscout-data", "players.json")

with open(players_path) as f:
    players = json.load(f)

df_players = pd.DataFrame(players)
df_players.info()

# get all countries of the players
df_players.birthArea.apply(lambda x: x['name']).unique()

# get al belgian players
df_belgian_players = df_players[df_players.birthArea.apply(lambda x: x['name'] == 'Belgium')]
df_belgian_players.head()

# players with 'both' as foot
df_players_both_foot = df_players[df_players.foot == 'both']
df_players_both_foot
```

```python
# EVENTS DATA
events_path = os.path.join(notebook_dir.parent, "wyscout-data", "events", "events_England.json")

with open(events_path) as f:
    events = json.load(f)

df_events = pd.DataFrame(events)
df_events.eventName.unique()

# get all offside events
df_offside_events = df_events[df_events.eventName == 'Offside']

# get match id of the match with the most offside events
match_id_most_offside_events = df_offside_events.groupby('matchId').size().sort_values(ascending=False).index[0]

# get the match with the most offsides
df_match_most_offsides = df_matches[df_matches.wyId == match_id_most_offside_events]
df_match_most_offsides
```
